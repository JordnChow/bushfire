```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("meteorological.csv")
df = df[df['area'] > 0].dropna(subset=['area'])
df['area'] = np.log1p(df['area'])
columns_to_plot = df.columns[4:11]  # Adjust this range if needed
x = df['area']
# Set up the subplots
fig, axs = plt.subplots(len(columns_to_plot), 1, figsize=(8, 3 * len(columns_to_plot)), sharex=False)

# If only one column, wrap axs in a list
if len(columns_to_plot) == 1:
    axs = [axs]

# Plot each variable vs. area
for i, column in enumerate(columns_to_plot):
    axs[i].scatter(df[column], x, label=f"{column} vs area", color=f'C{i % 10}')
    axs[i].set_title(f"{column} vs area burnt")
    axs[i].set_xlabel(column)
    axs[i].set_ylabel('Area Burnt')
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
